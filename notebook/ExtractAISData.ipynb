# extract AIS data from TVS

```python
__author__ = 'Administrator'

# To extract data points from big text files and save back with AIS structure

import time
import sys
import gc
inFile= 'F:/2015/201512/201512.tsv'
outPath = 'F:/2015/201512/'
# number of files
# number = 44
# readSize = 1024*1000*1000*2
#inFile= '//10.72.38.79/ais/201403.tsv'
#outPath = 'D:/share/201403/'
#number = 189
#readSize = 1024*1000*1000*10
number = 193
readSize = 1024*1000*1000*1

count_list = list(range(1,number+1))
expectedCols=18
with open(inFile,'r',errors='ignore') as readFile:
    lastStr=''

    for count in count_list:
        print(time.strftime("%Y-%m-%d %X",time.localtime()))
        s=readFile.read(readSize)
        with open(outPath+str(count)+'.txt','w') as wf:
            points=s.split('\n')
            #print(len(points))
            print(count)
            
            countNum = 0
            for point in points[1:-1]:
                cols=point.split('\t')
               
                aisKey=cols[1].split(' ')
                aisValue=cols[3].split('@')
                if len(aisKey)+len(aisValue)==expectedCols:
                    
                    aisValue=aisValue[0:8]
                    del aisValue[1]
                    del aisValue[2]

                    line=aisKey+aisValue
                    lineStr='@'.join(line)+'\n'
                    #print(lineStr)
                    wf.write(lineStr)
                else:
                    countNum=countNum+1
                    # print('1')
            print(countNum)
            wf.flush()
            del s
            del points
            gc.collect()

        print(time.strftime("%Y-%m-%d %X",time.localtime()))
```
